```python
import json
import numpy as np
#from sympy import *
init_printing(use_latex=True)
import scipy.spatial.distance as ssd
import matplotlib.pyplot as plt
%matplotlib inline
from pprint import pprint
    
def jsonInterpretor(filename):
    
    with open(filename) as f:
        data = json.load(f)          
  
    beats = 300 #the number of beats
    notes = "_notes"
    time = "_time"
    lineIndex = "_lineIndex" #row
    lineLayer = "_lineLayer" #column
    types = "_type" #0 is one color and 1 is the other
    cutDirection = "_cutDirection"#9 cut directions

    MatrixRows = beats * 8

    FinalMatrix = np.zeros(shape=(MatrixRows,32))#will hold all of the vectors, should be 32 by the number of beats * 8

    j = 0; #index for .json notes
    beat = 0.0 #1/8 increments for keeping track of time

    #data[notes][3][time]
    for i in range(MatrixRows):
        if data[notes][j][time] == beat:
            if data[notes][i][types] == 0: #if the color is equal to 0, start at zero
                if data[notes][i][lineIndex] != 0:
                    FinalMatrix[i][data[notes][i][lineIndex]]+=1
                if data[notes][i][lineLayer] != 0:
                    FinalMatrix[i][data[notes][i][lineLayer]+4]+=1
                if data[notes][i][cutDirection] != 0:
                    FinalMatrix[i][data[notes][i][cutDirection]+7]+=1
            else: #else the color is equal to 1, add 16
                if data[notes][i][lineIndex] != 0:
                    FinalMatrix[i][data[notes][i][lineIndex]+16]+=1
                if data[notes][i][lineLayer] != 0:
                    FinalMatrix[i][data[notes][i][lineLayer]+4+16]+=1
                if data[notes][i][cutDirection] != 0:
                    FinalMatrix[i][data[notes][i][cutDirection]+7+16]+=1
            j+=1#increment the note cntr after updating the matrix with the appropriate ones
        beat +=.25#increment the beat cntr after every 'i' increment
    
    return FinalMatrix;

```
